```python
#children_age<18 
#adult_age >=18
#male with children
#children with parents
#with sibling
#with spouse

[m,n]=data3.shape
temp_gender={'male':[0,0],'female':[0,0]}#classify either nake or female survival

temp_age={"female":{"withSibSp":{"adult":[0,0],"child":[0,0]},"withoutSibSp":{"adult":[0,0],"child":[0,0]},"withParch":{"adult":[0,0],"child":[0,0]},"withoutParch":{"adult":[0,0],"child":[0,0]}},"male":{"withSibSp":{"adult":[0,0],"child":[0,0]},"withoutSibSp":{"adult":[0,0],"child":[0,0]},"withParch":{"adult":[0,0],"child":[0,0]},"withoutParch":{"adult":[0,0],"child":[0,0]}}}
#classify baed on age and gender

for i in range(m):
    temp_gender[data3['Sex'][i]][data3['Survived'][i]]+=1
    rnge=''
    if data3["Age"][i]>=18:#18 is the age when I assume to be the starting point of adulthood
        rnge="adult"
    else:
        rnge="child"
    sibSp=""
    if data3["SibSp"][i]!=0:
        sibSp="withSibSp"
    else:
        sibSp="withoutSibSp"
    
    parch=""
    if data3["Parch"][i]!=0:
        parch="withParch"
    else:
        parch="withoutParch"
        
    temp_age[data3['Sex'][i]][parch][rnge][data3['Survived'][i]]+=1
    temp_age[data3['Sex'][i]][sibSp][rnge][data3['Survived'][i]]+=1

    
print(temp_gender,"\n\n")
print(temp_age)




    

    
```

```python
#calcualtion for evaluation

#total no of female
total_female=int(sum(temp_gender['female']))
#total no of survived female
survived_female=temp_gender['female'][1]
#total no of male
total_male=sum(temp_gender['male'])
#total no of male survive
survived_male=temp_gender['male'][1]
#percentage of female survive
percentage_female_survived=(survived_female/total_female)*100
#percentage of male survive
percentage_male_survived=(survived_male/total_male)*100
#percentage of female in survival people
percentage_female_from_survival=(survived_female/(survived_female+survived_male))*100
#percentage of people survive 
percentage_people_Survived=((survived_male+survived_female)/(total_female+total_male))*100



print(percentage_female_survived,percentage_male_survived,percentage_female_from_survival,percentage_people_Survived)
print("\n total no of passenger in titanic=\t",total_female+total_male,
     "\n \n columns>>\n total_count>> total male or female", "\n survived_count>> total no of female|male survived","\npercentage_of_total>> percentage among all the population"," \npercentage_survived>>> percentage of male|female survived among own gender ","\n percentage_among_survived>>> percentage of male|female among all the survived people"," \nsurvived_percentage_of_total>>>survived among all the population")


temp_dict={'gender':['male','female'],'total_count':[total_male,total_female],"survived_count":[survived_male,survived_female],'percentage_survived':[percentage_male_survived,percentage_female_survived]
         ,'survived_percent_of_total':[(survived_male/(total_male+total_female))*100,(survived_female/(total_male+total_female))*100] 
          ,"percentage_of_total":[100*(total_male/891),100*(total_female/891)]
           ,"percentage_among_survived":[100-percentage_female_from_survival,percentage_female_from_survival]
          }
df=pd.DataFrame.from_dict(temp_dict)

df
```

```python

df_age_dict={"class":["Total","not survive","survive","% survival","% survival_with_Parrents/child","% survival_without_Parrents/child","% survival_with_siblings/spouse","% survival_without_siblings/spouse"]}
for x in temp_age.keys():
    #adults
    
    #total no of adults
    total=int(sum(temp_age[x]["withParch"]["adult"])+sum(temp_age[x]["withoutParch"]["adult"]))
    #total of adults not survived
    notsurvive=temp_age[x]["withParch"]["adult"][0]+temp_age[x]["withoutParch"]["adult"][0]
    #total no of adults who survived
    survive=temp_age[x]["withParch"]["adult"][1]+temp_age[x]["withoutParch"]["adult"][1]
    #percentage of adults survive
    percentageSurvive=100*survive/total
    #percentage of survival who are with parents/chirldren
    percentage_survival_Parch=100*temp_age[x]["withParch"]["adult"][1]/total
    #percentage of  survival who are without parents/chirldren
    percentage_survival_withoutParch=100*temp_age[x]["withoutParch"]["adult"][1]/total
    #percentage of survival who are with siblings/spouse
    percentage_survival_SibSp=100*temp_age[x]["withSibSp"]["adult"][1]/total
    #percentage of  survival who are without siblings/spouse
    percentage_survival_withoutSibSp=100*temp_age[x]["withoutSibSp"]["adult"][1]/total
    
    #create the dictionary 
    temp_list=[total,notsurvive,survive,percentageSurvive,percentage_survival_Parch,percentage_survival_withoutParch,percentage_survival_SibSp,percentage_survival_withoutSibSp]
    temp_list= [ '%.2f' % elem for elem in temp_list ]
    df_age_dict[str(x)+"_adult"]=temp_list
    
    
    #child
    #total no of childs
    total=sum(temp_age[x]["withParch"]["child"])+sum(temp_age[x]["withoutParch"]["child"])
    #total no of childs not survive
    notsurvive=temp_age[x]["withParch"]["child"][0]+temp_age[x]["withoutParch"]["child"][0]
    #total no of childs  survive
    survive=temp_age[x]["withParch"]["child"][1]+temp_age[x]["withoutParch"]["child"][1]
    #percentage of child survival
    percentageSurvive=100*survive/total
    #percentage of survival who are with parents/chirldren
    percentage_survival_Parch=100*(temp_age[x]["withParch"]["child"][1]/total)
    #percentage of survival who are without parents/chirldren
    percentage_survival_withoutParch=100*(temp_age[x]["withoutParch"]["child"][1]/total)
    #percentage of survival who are with siblings/spouse
    percentage_survival_SibSp=100*(temp_age[x]["withSibSp"]["child"][1]/total)
    #percentage of survival who are without siblings/spouse
    percentage_survival_withoutSibSp=100*(temp_age[x]["withoutSibSp"]["child"][1]/total)
    
    
    temp_list=[total,notsurvive,survive,percentageSurvive,percentage_survival_Parch,percentage_survival_withoutParch,percentage_survival_SibSp,percentage_survival_withoutSibSp]
    temp_list= [ '%.2f' % elem for elem in temp_list ]
    
    df_age_dict[str(x)+"_child"]=temp_list
    #for y in temp_age[x]:
        #df_age_dict[str(x)+"_"+str(y)]=[sum(temp_age[x][y])]+temp_age[x][y]+[100*(temp_age[x][y][1])/sum(temp_age[x][y])]
df_age=pd.DataFrame(df_age_dict)
df_age
```

```python
#depending upon the passenger Id
def prime(num):
    if num > 1:
        for i in range(2,int(num/2)+1):
            if (num % i) == 0:
                return 0
        return 1
       
# if input number is less than
# or equal to 1, it is not prime
    else:
       return 0


temp_passId={2:{"count":0,'male':[0,0],'female':[0,0]},3:{"count":0,'male':[0,0],'female':[0,0]},5:{"count":0,'male':[0,0],'female':[0,0]},10:{"count":0,'male':[0,0],'female':[0,0]},"prime":{"count":0,'male':[0,0],'female':[0,0]}}

[m,n]=data3.shape

for i in range(m):
    #if passenger Id is divisible by 2
    if data3["PassengerId"][i]%2==0:
        temp_passId[2]["count"]+=1
        temp_passId[2][data3['Sex'][i]][data3['Survived'][i]]+=1
    
    #if passenger Id is divisible by 3
    if data3["PassengerId"][i]%3==0:
        temp_passId[3]["count"]+=1
        temp_passId[3][data3['Sex'][i]][data3['Survived'][i]]+=1
        
    #if passenger Id is divisible by 5
    if data3["PassengerId"][i]%5==0:
        temp_passId[5]["count"]+=1
        temp_passId[5][data3['Sex'][i]][data3['Survived'][i]]+=1
    
    #if passenger Id is divisible by 10
    if data3["PassengerId"][i]%10==0:
        temp_passId[10]["count"]+=1
        temp_passId[10][data3['Sex'][i]][data3['Survived'][i]]+=1
    
    #if passengerId is a prime no
    if prime(data3["PassengerId"][i]):
        temp_passId["prime"]["count"]+=1
        temp_passId["prime"][data3['Sex'][i]][data3['Survived'][i]]+=1
        
temp_passId     

```
